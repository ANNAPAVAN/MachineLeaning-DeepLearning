```python
# DISTPLOT (Numerical - Ctegorical)

sns.distplot(titanic["Age"])
```

```python
# HEATMAP (Categorical - Categorical)
titanic.head()
```

```python
pd.crosstab(titanic["Pclass"],titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
titanic.groupby("Pclass").mean()["Survived"]*100
# Calculating % of people survived
```

```python
(titanic.groupby("Pclass").mean()["Survived"]*100).plot(kind="bar")
```

```python
(titanic.groupby("Sex").mean()["Survived"]*100).plot(kind="bar")
```

```python
(titanic.groupby("Sex").mean()["Survived"]*100).plot(kind="pie")
```

```python
titanic.groupby("Embarked").mean()["Survived"]*100
```

```python
(titanic.groupby("Embarked").mean()["Survived"]*100).plot(kind="pie")
```

```python
# ClusterMap  (Categorical - Categorical)

pd.crosstab(titanic["SibSp"],titanic["Survived"])
```

```python
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```

```python
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```
