```python
datagen = ImageDataGenerator(
    rotation_range=30,      # Rotation angle (0-30 degrees)
    shear_range=0.2,        # Shear transformation
    zoom_range=0.2,         # Zoom in/out
    horizontal_flip=True,   # Horizontal flip
    width_shift_range=0.2,  # Horizontal shift
    height_shift_range=0.2  # Vertical shift
)
```

```python
from tensorflow.keras.utils import load_img, img_to_array
```

```python
img = img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) # 1 image processing here
```

```python
i=0
#first create aug folder
for output in datagen.flow(input_batch,batch_size=1,save_to_dir="aug"):
  i=i+1

  if i==10:
    break
```

```python
from IPython.display import Image, display
import os

# Specify the directory where your images are located
directory = "/content/aug"  # Replace with the name of your directory

# List all files in the directory
files = os.listdir(directory)

# Filter for image files (you can add more extensions if needed)
image_files = [f for f in files if f.lower().endswith((".jpg", ".jpeg", ".png", ".gif"))]

# Display the images
for file_name in image_files:
    # Create a full path to the image file
    file_path = os.path.join(directory, file_name)

    # Display the image
    display(Image(filename=file_path))
```
