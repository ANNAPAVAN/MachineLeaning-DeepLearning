```python
from sklearn.model_selection import train_test_split
```

```python
# x_train,x_test,y_train,y_test = train_test_split(df.drop("Purchased",axis=1),df["Purchased"],test_size=0.3,random_state=0)
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

# fit the scaler to train set , it will learn the parameters

scaler.fit(x_train)

# transforms train and test sets

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_  #provides age and salary mean
```

```python
x_train
```

```python
x_train_scaled  # now in numpy array form , we should convert into DataFrame
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
x_train_scaled
```

```python
np.round(x_train.describe(),1) #Original Dataset
```

```python
np.round(x_train_scaled.describe(),1) #Scaled Dataset  herre we see Mean=0 Std_Dev=1
```

```python
# WHY SCALING IS IMPORTANT
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred_scaled = lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual : ",accuracy_score(y_test,y_pred))
print("Scaled : ",accuracy_score(y_test,y_pred_scaled))
```
